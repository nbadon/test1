#### Ideas Scoring

```python
df[["I can buy/sell all Altcoins at \"P2P exchange\"..1"]].value_counts()
```

```python
def get_score(question):
    print(question)
    sum_of_score = sum(df[[question]].value_counts())
    return sum(df[[question]].value_counts().index.to_series().apply(lambda x: int(x[0])) 
               * df[[question]].value_counts())/sum_of_score
```

```python
get_score("I can buy/sell all Altcoins at \"P2P exchange\"..1")
```

```python
df[["I can \"Swap\" from BTC / ETH / UST to all Altcoins and vice versa..1"]].value_counts()
```

```python
get_score("I can \"Swap\" from BTC / ETH / UST to all Altcoins and vice versa..1")
```

```python
df[["I can \"Swap\" from my fiat-currency to all Altcoins and vice versa..1"]].value_counts()
```

```python
get_score("I can \"Swap\" from my fiat-currency to all Altcoins and vice versa..1")
```
